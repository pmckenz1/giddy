### MCMC tree sampling with decay

```python
import toytree
import toyplot
import numpy as np
import pandas as pd
import strange
```

```python
# load MCMC object with data for imb5
mcmc = strange.MCMC("imb5", "../tests/")
```

```python
# run MCMC mixing
mcmc.run(nsteps=100000, nsamps=1000)
```

```python
# plot likelihood over sampled iters
toyplot.plot(
    mcmc.score, 
    height=300,
    ylabel="-log-likelihood", 
    xlabel="MCMC sample",
);
```

```python
import strange
```

```python
# load all results and compare to true trees
comp = strange.Compare("imb5", "../tests")
```

```python
# run comparisons
comp.run()
```

```python
# look at rf distances for each method
comp.rf_table.iloc[:, 4:].mean()
```
